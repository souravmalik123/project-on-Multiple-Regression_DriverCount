```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
print(data.head(2))
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
print(data.describe().T)
```

```python
print(data.isnull().sum())
```

```python
sns.countplot(x=data["season"],data=data)
plt.show()
```

```python
sns.countplot(x=data["holiday"],data=data)
plt.show()
```

```python
sns.countplot(x=data["season"],hue = data["holiday"],data=data)
plt.show()
```

```python
sns.countplot(x=data["season"],hue = data["workingday"],data=data)
plt.show()
```

```python
sns.countplot(x=data["season"],hue = data["weather"],data=data)
plt.show()
```
